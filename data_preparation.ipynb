```python
import pandas as pd

df = pd.read_csv("SYB66_317_202310_Seats_held_by_women_in_Parliament.csv", encoding='latin1')
df
```

```python
# Drop rows where 'Region/Country/Area' contains specific strings, that give redundant information
df = df[~df['Region/Country/Area'].str.contains("Total, all countries or areas")]
df = df[~df['Region/Country/Area'].str.contains("LDC§")]

# Display the first few rows of the filtered data
df
```

```python
columns_to_drop = ['Index',
    'Last Election Date',
    'Last Election Date footnote',
    'Footnotes',
    'Source'
]

df = df.drop(columns=columns_to_drop)
df
```

```python
regions = [
    'Northern Africa', 'Sub-Saharan Africa', 'Eastern Africa', 'Southern Africa', 'Western Africa',
    'Northern America', 'Latin America & the Caribbean', 'Central America', 'South America',
    'Central Asia', 'Eastern Asia', 'South-eastern Asia', 'Southern Asia', 'Western Asia',
    'Caucasus and Central Asia', 'Eastern Europe', 'Northern Europe', 'Southern Europe', 'Western Europe', 'Oceania'
]

regions_data = df[df['Region/Country/Area'].isin(regions)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(14, 7))
latest_year = regions_data['Year'].max()
sns.barplot(data=regions_data[regions_data['Year'] == latest_year], x='Region/Country/Area', y='Value')
plt.title(f'Women Representation in Parliaments by Region ({latest_year})')
plt.xticks(rotation=90)
plt.ylabel('Percentage')
plt.xlabel('Region')
plt.show()
```

```python
df_sliced = df.iloc[250:]
df_sliced.head(10)
```

```python
# Save Regions Data
regions_data.to_csv('regions_data.csv')
```

```python
# Load country coordinates dataset
coordinates_df = pd.read_csv("average-latitude-longitude-countries.csv")

# Merge datasets based on country names
merged_df = pd.merge(df_sliced, coordinates_df[['Country', 'Latitude', 'Longitude']], 
                     left_on='Region/Country/Area', right_on='Country', how='left')

# Drop unnecessary columns and rename if needed
merged_df.drop(columns=['Country'], inplace=True)  # Drop duplicate column if merged on 'Country'
merged_df.rename(columns={'Latitude': 'Latitude', 'Longitude': 'Longitude'}, inplace=True)

# Save updated dataset with coordinates
merged_df.to_csv("cleaned_df.csv", index=False)

print("Coordinates added and saved successfully.")
```
